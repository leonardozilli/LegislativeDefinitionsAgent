```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
defs = pl.read_csv("../data/definitions_corpus/definitions.csv")
defs
```

```python
defs.filter(pl.col("definiens").is_null())
```

```python
# we remove definitions with less than 4 words and more than 500 words
defs = defs.with_columns(pl.col('definition_text').str.split(' ').list.len().alias('word_count')).sort('word_count').filter(pl.col('word_count') > 3).filter(pl.col('word_count') < 500)
defs
```

```python
describe = defs.select('word_count').describe()
describe
```

```python
defs.plot.hist('word_count', bins='auto')
```

```python
mean = describe.filter(pl.col('statistic') == 'mean').select('word_count').item()
std = describe.filter(pl.col('statistic') == 'std').select('word_count').item()
print(std)
```

```python
import numpy as np

word_lengths = defs['word_count']

# Calculate the mean and standard deviation
mean = describe.filter(pl.col('statistic') == 'mean').select('word_count').item()
std_dev = describe.filter(pl.col('statistic') == 'std').select('word_count').item()
# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(word_lengths, bins="auto", edgecolor='none')

# Add vertical lines for mean and standard deviation
plt.axvline(mean, color='r', linestyle='-', label='Mean')
plt.axvline(mean - std_dev, color='b', linestyle='--', label='1 Std. Dev.')
plt.axvline(mean + std_dev, color='b', linestyle='--')
plt.xlim(0, max(word_lengths) + 10)

# Customize the plot
plt.title('Definition Word Length Distribution')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
